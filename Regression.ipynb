# 1.Correlation Coefficient

```python
x = np.arange(10)
y = 2 * x + 1
print(f"x: {x}, y: {y}")

plt.scatter(x, y)
plt.title("Scatter plot of x vs y")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

피어슨 상관계수
- 선형(linear) 관계 나타냄
- 연속형(정규성 가정 필요)
- outlier에 민감
- 값 자체의 관계를 분석
- Ex) 시험 점수와 운동 실력 점수 간의 실제 값이 얼마나 선형적인지 계산

스피어만 상관계수
- 단조(monotonic) 관계 나타냄  *단조: 한 변수의 값이 증가할 때 다른 변수도 증가하거나 감소하는 방향이 일관되게 유지되는 관계
- 연속형 또는 서열형(비모수적)
- outlier에 덜 민감
- 순위 관계를 분석
- 점수 자체보다는 순위(1위, 2위, 3위 등) 간의 관계를 계산

결론
- 선형 관계 분석 → 피어슨
- 연속형 데이터, 정규성 가정 가능 → 피어슨
- 비선형(단조) 관계 분석 → 스피어만
- 이상치 영향 적게 받고 싶을 때 → 스피어만

```python
from scipy.stats import pearsonr, spearmanr

pearson_corr, pearson_p_value = pearsonr(x, y)

spearman_corr, spearman_p_value = spearmanr(x, y)

print("Pearson corr coef:", pearson_corr, "Pearson pvalue:", pearson_p_value)
print("Spearman corr coef:", spearman_corr, "Spearman pvalue:", spearman_p_value)
#pvalue < 0.05: 상관관계가 있다.
```

해석
- Pearson Correlation Coefficient 높다 -> 양의 선형관계
- Spearman Correlation Coefficient 높다 -> 단조 증가관계(선형적인지는 모름)
- Pearson P-value < 0.05 -> 통계적으로 유의미한 상관관계가 있음
- Spearman P-value < 0.05 -> 순위(랭크)간에 상관관계가 있음
